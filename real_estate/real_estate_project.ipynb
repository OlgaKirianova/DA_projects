## Предобработка данных

### Обработка пропусков


```python
data.isna().sum()
```

Первые пропуски встречаются в колонке `ceiling_height`, и это неудивительно - продавец либо не смог измерить или найти информацию о высоте потолка по типу серии дома, либо не посчитал нужным указывать. Вообще на ресурсах по недвижимости в описании объекта этот параметр часто опускают. Пропуски составляют целых 38% - это очень большая часть данных. Их можно было бы заполнить медианным значением, используя колонку с количеством этажей, но это было бы не корректно, так как 5-этажный дом может оказаться как панельным с минимальной высотой или старым фондом с четырехметровыми потолками. С другой стороны, в Петербурге квартиры с высокими потолками - либо элитное жилье бизнес-класса и выше, либо сталинки/старый фонд, такие типы домов крайне редко бывают выше 12 этажей. Поэтому для объектов с количеством этажей выше этой цифры можно принять медианное по этажности, так как большинство окажется типовыми сериями со стандартной высотой потолков. 

```python
data.loc[data['floors_total'] > 12, 'ceiling_height'] = data[
    data['floors_total'] > 12].groupby(
    'floors_total')['ceiling_height'].apply(lambda x: x.fillna(x.mean()))

data['ceiling_height'] = data['ceiling_height'].round(2)
# округлим данные о высоте потолков до сотых

data['ceiling_height'].isna().sum()
# посмотрим сколько пропусков осталось незаполненными
```

Пропуски в `living_area` можно заполнить медианными значениями на основе группировки по общей площади.

Пропуски в `kitchen_area` закономерны, если объект - студия, так как в этом случае площадь кухни не входит в жилую площадь квартиры и при определении «студии» не учитывается, эти пропуски можно заполнить нулем. Остальные - медианным значением, также используя данные общей площади.  

```python
data['living_area'] = data.groupby(
    'total_area')['living_area'].apply(lambda x: x.fillna(x.median()))
# заполняем пропуски в жилой площади 

data.loc[data['studio'] == True, 'kitchen_area'] = data.loc[
    data['studio'] == True, 'kitchen_area'].fillna(0) 
# заполняем пропуски для квартир-студий

data['kitchen_area'] = data.groupby(
    'total_area')['kitchen_area'].apply(lambda x: x.fillna(x.median()))
# заполняем пропуски площади кухни 
```

Пропуски в столбце `is_apartment` разумно заменить на False. Обычно этот параметр указывают только тогда, когда недвижимость имеет статус апартаментов, в противном случае его скорей всего просто игнорируют. Поэтому примем, что такие пропуски - это квартиры. 

```python
data['is_apartment'] = data['is_apartment'].fillna(False) 
# заполняем пропуски в апартаментах
```

Пропуски в `balcony` также логичны и определенно указывают на отсутствие балкона, заменим на ноль, а заодно приведем к целочисленному типу, ведь не может быть полтора балкона.

В `locality_name` пропущено всего 49 строк, это совсем мало, их можно заменить на "unknown", так как данные о названиях населенных пунктов невозможно получить на основании других столбцов.

```python
data['balcony'] = data['balcony'].fillna(0).astype('int32') 
# заполняем пропуски балконов нулем, приводим к целочисленному 

data['locality_name'] = data['locality_name'].fillna('unknown') 
# заполняем пропуски в названиях населенных пунктов
```


Пропуски в `days_exposition` говорят о том, что на момент сбора данных объявление не было снято с публикации, а недвижимость еще продавалась, поэтому эту колонку мы не трогаем.

Также мы видим достаточное количество пропусков во всех столбцах, полученных автоматически на основе картографических данных. Возможно, причина в ошибке при выгрузке или другом техническом сбое. ## Расчёты и добавление результатов в таблицу

Добавим в таблицу цену за квадратный метр и переведем в тип *int* так как копейки не играют никакой роли и на сайтах недвижимости не указываются.

```python
data['price_meter'] = (data['last_price'] / data['total_area']).astype('int32')

# добавляем расчет с ценой за квадратный метр, переводим в целочисленный тип
```

Добавим столбцы с днем недели, месяцем и годом публикации объявления


```python
data['year_exposition'] = data['first_day_exposition'].dt.year

data['month_exposition'] = data['first_day_exposition'].dt.month

data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
```

Добавим столбец с категорией этажа квартиры — *первый, последний, другой*. Этот параметр важен для дальнейшего анализа и оценки, так как часто выставляется в фильтрах при поиске недвижимости покупателем.

```python
def floor_type(row):  # функция, которая на основе данных об этажности объекта определит категорию
 
    floor = row['floor']
    floors_total = row['floors_total']  
 
    if floor == 1:
        return 'первый'
    if floors_total == floor:
        return 'последний'
    return 'другой'
 
row_values = [2,4]
row_columns = ['floor', 'floors_total']
row = pd.Series(data=row_values, index=row_columns) 
floor_type(row) 

data['floor_type'] = data.apply(floor_type, axis=1) # добавляем колонку с категорией этажа
```

Добавим столбцы с соотношением жилой и общей площади, а также отношение площади кухни к общей.

```python
data['ratio_total_living'] = (data['living_area'] / data['total_area']).round(2)

# считаем соотношение жилой и общей площади, округляем до сотых

data['ratio_total_kitchen'] = (data['kitchen_area'] / data['total_area']).round(2)

# считаем соотношение общей площади и площади кухни, округляем до сотых
```
